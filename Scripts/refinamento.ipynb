```python
import duckdb
```

```python
con = duckdb.connect(database='dados_duckdb.db',read_only=False)
```

```python
df = con.execute("SELECT * FROM produtos").fetchdf()
df.head()
```

```python
df2 = df.drop(columns=['Categoria','Fornecedor'])
df2.head(10)
```

```python
con.execute("""
     DROP TABLE dim_produtos
""")
```

```python
con.execute("""
    CREATE TABLE IF NOT EXISTS dim_produtos(
    id_produto BIGINT,
    Nome_produto TEXT,
    Preco_produto FLOAT
    )
""")
```

```python
con.execute("INSERt INTO dim_produtos SELECT * FROM df2")
```

```python
df_Dim = con.execute("SELECT * FROM dim_produtos").fetchdf()
df_Dim.head()
```

```python
con.close()
```
